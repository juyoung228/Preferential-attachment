## Generating the similarity matrix for the structures

```python
#Import the required libraries
from rdkit import Chem
from rdkit import DataStructs
from rdkit.Chem.Fingerprints import FingerprintMols
from rdkit.Chem.AtomPairs import Pairs
from rdkit.Chem.AtomPairs import Torsions
```

```python
# Read the input file and extract each line into an array
with open('../cleaned_drug_struct.txt') as file: 
    lines = file.read().splitlines()
```

```python
# extract the structures and their respective names from the file
drug_names_from_file = [] # this contains invalid structures; we remove those and place them in another list later
drug_struct = []

for line in lines:
    split_line = line.split('\t')
    drug_names_from_file.append(split_line[1])
    drug_struct.append(split_line[0])
```

```python
# Extract the molecular structure from the smiles text
molecular_structures = []

for struct in drug_struct:
    molecular_structures.append(Chem.MolFromSmiles(struct))
```

```python
# Convert the molecular structures into fingerprints to be compared via the Tanimoto distance
fingerprints = []
drug_names = [] # This is a list of all of the valid drug names
invalid_drug_structures = [] # This is a list of the invalid drug names + structures.

for molecular_structure, drug_name in zip(molecular_structures, drug_names_from_file):
    try:
        fingerprints.append(FingerprintMols.FingerprintMol(molecular_structure))
        drug_names.append(drug_name)
    except:
        invalid_drug_structures.append({drug_name: molecular_structure})
```

```python
# Calculate and generate the similarlity matrix
similarity_matrix = [] 

# Tanimoto, Dice, Cosine, Sokal, Russel, Kulczynski, McConnaughey, and Tversky.
similarity_metric = DataStructs.DiceSimilarity

for structure_outer, drug_name_outer in zip(fingerprints, drug_names):
    matrix_row = []
    for structure_inner, drug_name_inner in zip(fingerprints, drug_names):
        similarity = DataStructs.FingerprintSimilarity(structure_outer, structure_inner, metric=similarity_metric)
        matrix_row.append(similarity)
    similarity_matrix.append(matrix_row)
```

## Heatmap of the Similarity matrix

```python
import matplotlib.pyplot as plt

plt.imshow(similarity_matrix, cmap='hot', interpolation='nearest')
plt.show()
```


## Boxplot of the similarity values

```python
import numpy as np

# Convert similarity_matrix to numpy matrix
matrix = np.array(similarity_matrix)

# We only need the upper or lower triangle; the unique comparisons
indices = np.triu_indices_from(matrix)
matrix = np.asarray( matrix[indices] )

plt.boxplot(matrix)
plt.show()
```

## Write the results to a file

```python
# Write the results to a csv file
import csv 

file_path = '../struct/struct_sim_tanimoto.txt'
with open(file_path, 'w') as out:
    writer = csv.writer(out)
    writer.writerows(similarity_matrix)
```

## Write the results to a file with labels

```python
# Adding an empty value at the front to correctly position the headers
headers = [' '] + drug_names

modified_sim_matrix = []

for i in range(len(similarity_matrix)):
    modified_sim_matrix.append([drug_names[i]] + similarity_matrix[i])

modified_sim_matrix = [headers] + modified_sim_matrix

file_path = '../struct/tanimoto_with_labels.txt'
with open(file_path, 'w') as out:
    writer = csv.writer(out)
    writer.writerows(modified_sim_matrix)
```
